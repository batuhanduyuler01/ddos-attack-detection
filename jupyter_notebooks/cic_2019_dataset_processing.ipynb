```python
def printDataStatistics(dataset: pd.DataFrame) -> None :
    print(f'No of Labels: {len(dataset[" Label"].unique())}')
    print(dataset[" Label"].unique())
    
    
    ddosPaketSayisi = len(dataset[dataset[" Label"] != "BENIGN"])
    normalPaketSayisi = len(dataset[dataset[" Label"] == "BENIGN"])
    paketNumber = len(dataset)
    
    
    print("\n---DataSet---\n")
    print(f'Atak Sayısı: {ddosPaketSayisi}')
    print(f'Normal Paket Sayısı: {normalPaketSayisi}')
    print(f'Total Sayısı: {paketNumber}')
    print(f'Atak Oranı: {100 * ddosPaketSayisi / paketNumber}')
```

```python
veriSetiListesi : list = [synDataSet, udpLagDataSet, ntpDataSet]
```

```python
for data in veriSetiListesi : 
    printDataStatistics(data)
```

```python
def __clean_data_set(dataframe: pd.DataFrame) -> pd.DataFrame:
    dataframe.dropna(inplace=True)
    indicesToKeep = ~dataframe.isin([np.nan, np.inf, -np.inf]).any(1)
    return dataframe[indicesToKeep].astype(np.float64)
```
